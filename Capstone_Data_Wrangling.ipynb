```python
df['review_score_totals'] = (df['review_scores_accuracy'] + df['review_scores_cleanliness'] + df['review_scores_checkin'] + df['review_scores_communication'] + df['review_scores_location'] + df['review_scores_value'])/6
```

```python
df['recent_availability'] = (df['availability_30'] + df['availability_60'] + df['availability_90'])/ df['availability_365']
newra = []
n = float(180/365)

for i in df['recent_availability']:
    if i == n:
        i = 0 #listing is free for the past 365 days
    elif i < n:
        i = 1 #listing is booked more often in the recent 90 days
    elif i > n:
        i = -1 #listing is unbooked for the past 365 days if the number is large or not recently booked
    elif pd.isnull(i):
        i = 1 #listing is booked for the past 365 days
    newra.append(i)
    
df['recent_availability'] = newra
```

```python
new_col = []

for i in df['host_response_time']:
    if i == 'within an hour':
        i = 1
    elif i == 'within a few hours':
        i = 2
    elif i == 'within a day':
        i = 3
    elif i == 'a few days or more':
        i = 4
    else:
        i = 0
    new_col.append(i)

df['host_response_time'] = new_col
```

```python
df.calendar_updated.unique()
```

```python
new_col = []

for i in df['calendar_updated']:
    if 'never' in i:
        i = -1
    elif 'today' in i:
        i = 0
    elif 'yesterday' in i:
        i = 1
    elif 'days ago' in i:
        i = re.search(r'\d+', i).group()
    elif 'a week ago' in i:
        i = 7
    elif 'weeks ago' in i:
        i = 7 * int(re.search(r'\d+', i).group())
    elif 'months ago' in i:
        i = 30 * int(re.search(r'\d+', i).group())
    new_col.append(i)
df['calendar_updated'] = new_col
```

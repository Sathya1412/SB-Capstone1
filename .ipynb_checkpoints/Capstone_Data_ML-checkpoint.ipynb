```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
from scipy.stats import norm
import scipy.stats as stats
import math

pd.set_option('display.height', 2000)
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 2000)
pd.set_option('display.width', 2000)
sns.set(font_scale = 1.5)
```

```python
df = pd.read_csv('listings_cleaned.csv')
```

```python
dff = df[['popular', 'id', 'name', 'host_name', 'host_since', 'availability_30', 'availability_60', 'availability_90', 
          'availability_365', 'number_of_reviews', 'accommodates', 'bedrooms', 'reviews_per_month',
         'neighbourhood', 'price', 'weekly_price', 'monthly_price', 'security_deposit', 'cleaning_fee',
         'amenities', 'bathrooms', 'beds', 'host_verifications', 'summary_len', 'description_len', 'transit_has',
         'no_smoking', 'no_pets', 'host_response_time_a few days or more', 'host_response_time_within a day',
         'host_response_time_within a few hours', 'host_response_time_within an hour']]
```

## Count number of NaNs for each column

```python
dff.isna().sum()
```

### Drop nans from bedrooms, neighbourhood, bathrooms, and beds. 
Can't drop nans from reviews_per_month because there are too many and we won't be using reviews_per_month anyway

```python
dff = dff.dropna(axis=0, subset=['bedrooms', 'neighbourhood', 'bathrooms', 'beds'])
```

```python
dff['popular'] = dff['popular'].apply(lambda x: 1 if x else 0)
dff['no_pets'] = dff['no_pets'].apply(lambda x: 1 if x else 0)
dff['no_smoking'] = dff['no_smoking'].apply(lambda x: 1 if x else 0)
```

```python
dff['price'] = dff['price'].str.replace(',', '')
dff['price'] = dff['price'].str.replace('$', '')
dff['price'] = dff['price'].astype(float)
```

## Set features dataframes and target so KNeighborsClassifier can be used
Explanation: There are 8 features dataframes. Each dataframe has similar columns. 
- features: Contains all columns
- neifeatures: Contains neighborhoods
- avafeatures: Contains all availabilities
- accfeatures: Contains all objects/etc present in the house/listing
- desfeatures: Contails summaries and descriptions that were provided by the host
- rulfeatures: Contains rules and policies 
- hosfeatures: Contains host response times
- prifeatures: Contains price
- target: Target column of the KNN Classifier, in this case its 'popular'

```python
features = dff[['availability_30', 'availability_60', 'availability_90', 'availability_365', 'accommodates', 
                'bedrooms', 'neighbourhood', 'price', 'bathrooms', 'beds', 'summary_len', 'description_len', 
                'transit_has', 'no_smoking', 'no_pets', 'host_response_time_a few days or more', 
                'host_response_time_within a day', 'host_response_time_within a few hours', 
                'host_response_time_within an hour']]

neifeatures = dff[['neighbourhood']]

avafeatures = dff[['availability_30', 'availability_60', 'availability_90', 'availability_365']]

accfeatures = dff[['accommodates', 'bedrooms', 'bathrooms', 'beds']]

desfeatures = dff[['summary_len', 'description_len']]

rulfeatures = dff[['transit_has', 'no_smoking', 'no_pets']]

hosfeatures = dff[['host_response_time_a few days or more', 'host_response_time_within a day', 
            'host_response_time_within a few hours', 'host_response_time_within an hour']]

prifeatures = dff[['price']]

target = dff['popular']
```

Each of the above dataframes are tested for their KNN score accuracies. This way I can check if there is any major differences between them. The review columns were not included since popular was a feature created using number of reviews as an indicator.

Optimal k is chosen from the graph

```python
features = pd.get_dummies(features, columns=['neighbourhood'])
neifeatures = pd.get_dummies(neifeatures, columns=['neighbourhood'])
```

```python
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.model_selection import train_test_split
```

```python
X = features.values
y = target.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42, stratify=y)

neighbors = np.arange(1, 10)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)

plt.title('AllFeatures k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(range(10))
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
```

## Lasso is used to predict which features influence popularity the most

```python
X = features.values
y = target.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=.0001, normalize=True)
col_names = features.columns
lasso_coef = lasso.fit(X, y).coef_
print(lasso_coef)

plt.figure(figsize=(20, 10))
plt.plot(range(len(col_names)), lasso_coef)
plt.ylabel('Influence on Popularity')
plt.xlabel('Features')
plt.title('Influence vs Features')
plt.xticks(range(len(col_names)), col_names, rotation=90)
plt.margins(0.02)
plt.show()
```

## Classification Report and Confusion Matrix
Precision, recall, f1-score are very important to know when working with machine learning models such as KNN

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

X = features.values
y = target.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42, stratify=y)

knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))

y_pred = knn.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## ROC curve, ROC AUC score, and cross val score
ROC AUC shows how good the classifier is at predict true and false positive rates. They are calculated using the cross val score. The results of the cross validation are acceptable since they are mostly around 70%.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score

y_pred_prob = knn.predict_proba(X_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

cv_auc = cross_val_score(knn, X, y, cv=5, scoring='roc_auc')

print("False Positive Rates = {}\n\nTrue Positive Rates = {}\n\nThresholds = {}\n".format(fpr, tpr, thresholds))
print("AUC scores computed using 5-fold cross-validation: {}".format(cv_auc))

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

## GridSearchCV
Grid search is used to determine the best value for the n_neighbors in KNN Classifier. 

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': np.arange(1,10)}

knn_cv = GridSearchCV(knn, param_grid, cv=5)

knn_cv.fit(X, y)

print("Tuned Logistic Regression Parameters: {}".format(knn_cv.best_params_)) 
print("Best score is {}".format(knn_cv.best_score_))
```

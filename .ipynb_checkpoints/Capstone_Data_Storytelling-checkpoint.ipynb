# Is there a linear relationship between the number of accommodates and the number of bedrooms?
Null Hypothesis: There is not a significant linear relationship between the number of accommodates and the number of bedrooms.
<br>Alternate Hypothesis: There is a significant linear relationship between the number of accommodates and the number of bedrooms.
### Correlation test

```python
sns.regplot(x='accommodates',y='bedrooms',data=df)
plt.show()
```

```python
df['accommodates'].isnull().value_counts()
```

```python
df['bedrooms'].isnull().value_counts()
```

```python
#There is only 1 null so replacing it with 0 won't change the result significantly
coef, pval = stats.pearsonr(df['accommodates'].fillna(0).tolist(), df['bedrooms'].fillna(0).tolist())
```

```python
coef
```

```python
pval
```

```python
df[['accommodates','bedrooms']].corr(method='pearson')
```

The p value < .05 so the null hypothesis is rejected. There is a significant linear relationship between the number of bedrooms and the number of accommodates.

# Does having lower prices make a listing more popular?
<p>Null hypothesis: Popular listings have the same prices as non-popular listings<br>
Alternate hypothesis: Popular listings don't have the same prices than non-popular listings</p>
### Two sample t-test

```python
prices = [float(x.translate(str.maketrans("", "", "$,"))) for x in pd.value_counts(df['price'][~df['popular']]).keys()]
p_prices = [float(x.translate(str.maketrans("", "", "$,"))) for x in pd.value_counts(df['price'][df['popular']]).keys()]
```

```python
stats.ttest_ind(p_prices, prices)
```

```python
prices_mean = np.mean(prices)
p_prices_mean = np.mean(p_prices)
```

```python
prices_mean
```

```python
p_prices_mean
```

Null Hypothesis is rejected since p value < .05, so there is a difference between the prices of popular listings and the prices of normal listings.

# Is not allowing smoking independent from being popular?
<p>Null hypothesis: A listing not allowing smoking and being popular are independent<br>
Alternate hypothesis: A listing not allowing smoking and being popular are not independent</p>
### Chi Square

```python
p = df['popular'].value_counts().tolist()
p
```

```python
ns = df['no_smoking'].value_counts().tolist()
ns
```

```python
NS_array = np.array(df['no_smoking'])
P_array = np.array(df['popular'])
```

```python
nsp = np.sum(NS_array & P_array) #no smoking and popular
nsnp = np.sum(NS_array & ~P_array) #no smoking and no popular
sp = np.sum(~NS_array & P_array) #smoking and popular
snp = np.sum(~NS_array & ~P_array) #smoking and no popular
st = sp + snp
nst = nsp + nsnp
pt = sp + nsp
npt = snp + nsnp
tt = st + nst
```

```python
print("           popular   not popular   total")
print("smoking    {}       {}          {}".format(sp, snp, st))
print("no smoking {}       {}          {}".format(nsp, nsnp, nst))
print("total      {}       {}          {}".format(pt, npt, tt))
```

```python
p_sp = sp/pt
p_nsp = nsp/pt
p_snp = snp/npt
p_nsnp = nsnp/npt
p_st = st/tt
p_nst = nst/tt
p_pt = p_sp + p_nsp
p_npt = p_snp + p_nsnp
p_tt = p_st + p_nst
```

```python
print("           popular   not popular   total")
print("smoking    {:.2%}     {:.2%}       {:.2%}".format(p_sp, p_snp, p_st))
print("no smoking {:.2%}     {:.2%}       {:.2%}".format(p_nsp, p_nsnp, p_nst))
print("total      {:.2%}    {:.2%}     {: .2%}".format(p_pt, p_npt, p_tt))
```

```python
ef_sp = (pt*st)/tt
ef_nsp = (pt*nst)/tt
ef_snp = (st*npt)/tt
ef_nsnp = (nst*npt)/tt
```

```python
f_obs = [sp, nsp, snp, nsnp]
f_obs
```

```python
f_exp = [ef_sp, ef_nsp, ef_snp, ef_nsnp]
f_exp
```

```python
stats.chisquare(f_obs, f_exp)
```

Since p value < .05, the null hypothesis is rejected. A listing not allowing smoking and being popular are not independent. This means that being a popular listing is related to whether or not smoking is allowed. 

# Is host response time independent from being popular?
<p>Null hypothesis: Host response time and being popular are independent<br>
Alternate hypothesis: Host response time and being popular are not independent</p>
### Chi Square

```python
p = df['popular'].value_counts().tolist()
p
```

```python
d3 = df['host_response_time_a few days or more'] 
d2 = df['host_response_time_within a day'] 
d1 = df['host_response_time_within a few hours'] 
d0 = df['host_response_time_within an hour'] 
d = d0 + d1+d1 + d2+d2+d2 + d3+d3+d3+d3 # 1 = within an hour, 2 = few hours, 3 = within a day, 4 = few days or more
d.value_counts().sort_index()
```

```python
(d0 + d1 + d2 + d3).value_counts()
```

```python
FD_array = np.array(df['host_response_time_a few days or more'])
D_array = np.array(df['host_response_time_within a day'] )
FH_array = np.array(df['host_response_time_within a few hours'])
H_array = np.array(df['host_response_time_within an hour'] )
P_array = np.array(df['popular'])
```

```python
hp = np.sum(H_array & P_array)
hnp = np.sum(H_array & ~P_array)
fhp = np.sum(FH_array & P_array)
fhnp = np.sum(FH_array & ~P_array)
dp = np.sum(D_array & P_array)
dnp = np.sum(D_array & ~P_array)
fdp = np.sum(FD_array & P_array)
fdnp = np.sum(FD_array & ~P_array)
ht = hp + hnp
fht = fhp + fhnp
dt = dp + dnp
fdt = fdp + fdnp
pt = hp + fhp + dp + fdp
npt = hnp + fhnp + dnp + fdnp
tt = ht + fht + dt + fdt
```

```python
print("          popular   not popular   total")
print("hour      {}       {}          {}".format(hp, hnp, ht))
print("few_hours {}       {}          {}".format(fhp, fhnp, fht))
print("day       {}        {}           {}".format(dp, dnp, dt))
print("few_days  {}         {}            {}".format(fdp, fdnp, fdt))
print("total     {}       {}          {}".format(pt, npt, tt))
```

```python
mx = [
    [hp, hnp, ht],
    [fhp, fhnp, fht],
    [dp, dnp, dt],
    [fdp, fdnp, fdt],
    [pt, npt, tt]
]
mx
```

```python
ex_mx = [
    [((mx[4][0]*mx[0][2])/mx[4][2]), ((mx[4][1]*mx[0][2])/mx[4][2])],
    [((mx[4][0]*mx[1][2])/mx[4][2]), ((mx[4][1]*mx[1][2])/mx[4][2])],
    [((mx[4][0]*mx[2][2])/mx[4][2]), ((mx[4][1]*mx[2][2])/mx[4][2])],
    [((mx[4][0]*mx[3][2])/mx[4][2]), ((mx[4][1]*mx[3][2])/mx[4][2])]
    ]
ex_mx
```

```python
f_obs = [mx[0][0], mx[0][1], mx[1][0], mx[1][1], mx[2][0], mx[2][1], mx[3][0], mx[3][1]]
f_obs
```

```python
f_exp = [ex_mx[0][0], ex_mx[0][1], ex_mx[1][0], ex_mx[1][1], ex_mx[2][0], ex_mx[2][1], ex_mx[3][0], ex_mx[3][1]]
f_exp
```

```python
stats.chisquare(f_obs, f_exp)
```

Since p value < .05, the null hypothesis is rejected. Host response time and being popular are not independent. This means that being a popular listing is related to how fast a host responds after you ask to rent the listing.
